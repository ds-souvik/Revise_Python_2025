```python
import numpy as np
```

### 1. Functions: power(), absolute(), sin(), cos(), log()

#### 1.a. power() function

```python
sample_array_1 = np.array([[1,2,-3,-4], [-5,-6,7,8]])

sample_array_1
```

```python
np.power(sample_array_1, 3)
```

```python
np.power(sample_array_1, np.array([0,1,2,3]))
```

```python
np.power(sample_array_1, np.array([[0,1,2,3], [0,0.5,0.75, 1.25]]))
```

#### 1.b. absolute() function

```python
np.absolute(sample_array_1)
```

#### 1.c. sin(), cos(): In numpy np.sin(30), 30 means 30 radians not 30 degrees. To convert rad to degrees = (radians * np.pi/180)

```python
np.sin(30 * np.pi/180)
```

#### 1. d. log()

```python
np.log(np.array([[1,2,3,4], [5,6,7,8]]))
```

#### 2. Empty and Multiply function

```python
y = np.empty((2, 4), dtype= 'int')

np.multiply(sample_array_1, 10, out = y)

y
```

### 3. Aggregate functions, Reduce and accumulate.

#### 3. a. reduce() functions

```python
sample_array_1
```

```python
np.add.reduce(sample_array_1)
```

```python
np.add.reduce(sample_array_1[1])
```

```python
np.multiply.reduce(sample_array_1)
```

```python
np.multiply.reduce(sample_array_1[1])
```

#### 3. b. accumulate(): Running total, running multiplication

```python
sample_array_1[1]
```

```python
np.add.accumulate(sample_array_1[1])
```

```python
np.multiply.accumulate(sample_array_1[1])
```
